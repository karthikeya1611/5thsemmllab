```python
import pandas as pd
import numpy as np

df = pd.read_csv('student_data.csv')

print("Mean Attendance: ", df['Attendance'].mean())
print("Avg Internal Marks: ", df['Internal Marks'].mean())
print("Avg External Marks: ", df['External Marks'].mean())
print("Max Internal Marks: ", df['Internal Marks'].max())
print("Min Internal Marks: ", df['Internal Marks'].min())
print("Max External Marks: ", df['External Marks'].max())
print("Min External Marks: ", df['External Marks'].min())
print("SD of Internal Marks: ", np.sqrt(df['Internal Marks'].var()))

print("\nSummary Statistics: ", df.describe())
```

```python
import pandas as pd

df = pd.read_csv('student_data.csv')

df['Total Marks'] = df['Internal Marks'] + df['External Marks']

print(df.to_string())

df['Percentage'] = (df['Total Marks'] / 100) * 100

print(df.to_string())


df.to_csv('new_student_data.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv('new_student_data.csv')

print("Students having attendence greater than 90 details: \n", df[df['Attendance'] > 90].to_string())
#print(df[df['Attendance'] > 90]['Name'])

print("Students having Total Marks greater than 90 details: \n", df[df['Total Marks'] > 90].to_string())
# df.loc[condition, 'Name']
#df[condition]['Name']
print("Female Students details: \n", df[df['Gender'] == ' F']['Name'].tolist())
print("Male Students details: \n", df[df['Gender'] == ' M']['Name'].tolist())

print("Students having External Marks less than 50 details: \n", df[df['External Marks'] < 50].to_string())
```

```python
import pandas as pd

df = pd.read_csv('new_student_data.csv')

print("Student having the max marks: ", df.loc[df['Total Marks'].idxmax()]['Name'])
print("Student having the min marks: ", df.loc[df['Total Marks'].idxmin()]['Name'])
print("Top 3 students based on Total Marks: \n", df.nlargest(3, 'Total Marks').to_string())
print("Bottom 3 students based on Total Marks: \n", df.nsmallest(3, 'Total Marks').to_string())
print("Number of people above average marks: ", len(df[df['Total Marks'] > df['Total Marks'].mean()]))
```

```python
import pandas as pd

df = pd.read_csv('new_student_data.csv')

print("Average total marks for Males: ", df[df['Gender'] == ' M']['Total Marks'].mean())
print("Average total marks for Females: ", df[df['Gender'] == ' F']['Total Marks'].mean())
print("Average Attendance: ", df['Attendance'].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('new_student_data.csv')

plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Total Marks'])
plt.show()

plt.hist(df['Attendance'])
plt.xlabel("Attendance")
plt.ylabel("Number of Students")
plt.show()

#x = [len(df[df['Gender'] == ' M']), len(df[df['Gender'] == ' F'])]
x = df["Gender"].value_counts()
plt.pie(x, labels=x.index)
plt.show()

plt.scatter(df['Attendance'], df['Total Marks'])
plt.xlabel('Attendance')
plt.ylabel('Total Marks')
plt.show()
```
